```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud
```

```python
data = pd.read_csv("Fixed_Dataset2.csv")
```

```python
data['name'] = data['name'].apply(word_tokenize)
```

```python
def combine_words(word_list):    
    all_words = []
    for word in word_list:
        all_words += word
    return all_words
```

```python
name = data['name']
words =  combine_words(name)
words[:10]
```

```python
mostcommon = FreqDist(words).most_common(50)
wordcloud = WordCloud(width=1600, height=800, background_color='white').generate(str(mostcommon))
fig = plt.figure(figsize=(30,10), facecolor='white')
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.title('Top 100 Most Common Words', fontsize=50)
plt.show()
```
